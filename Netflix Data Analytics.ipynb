## In which year were the highest number of contents released?


```python
# to check the data types
new_data.dtypes
```

```python
# creating a new dates column
new_data['Date_N'] = pd.to_datetime(new_data['Release_Date'])
```

```python
# to show the year counts
new_data['Date_N'].dt.year.value_counts()
```

## Which country released the highest number of Movies?

```python
# creating a new dataframe for movies only
data_movies = new_data[new_data['Category'] == 'Movie']
```

```python
data_movies.head()
```

```python
data_movies.Country.value_counts().head(10)
```

```python
plt.figure(figsize=(15,10))
sns.set(style='darkgrid')
ax = sns.countplot(y='Country', data=data_movies, palette="Set1", order=data_movies['Country'].value_counts().index[0:10])
```

United States released the highest no of movies amounting to 1627.

## Analysis of the duration of movies

```python
data_movies['Duration']=data_movies['Duration'].str.replace(' min','')
data_movies['Duration']=data_movies['Duration'].astype(str).astype(int)
data_movies['Duration']
```

```python
sns.set(style="darkgrid")
sns.kdeplot(data=data_movies['Duration'], shade=True)
```

So, a good amount of movies on Netflix are among the duration of 75-120 mins. It is acceptable considering the fact that a fair amount of the audience cannot watch a 3 hour movie in one sitting.
